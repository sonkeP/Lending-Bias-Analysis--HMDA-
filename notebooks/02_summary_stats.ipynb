```python
import glob
import pandas as pd

files = sorted(glob.glob("../data/clean/hmda/hmda_2023_parts/part_*.parquet"))

# Read a subset of parts for a manageable sample 
sample_files = files[:10]
df = pd.concat((pd.read_parquet(f) for f in sample_files), ignore_index=True)

df.to_parquet("../data/clean/hmda/hmda_2023_sample.parquet", index=False)
print("Sample rows:", len(df), "from parts:", len(sample_files))
```

```python
summary_gender = df.groupby("applicant_gender").agg(
    applications=("approved","size"),
    approval_rate=("approved","mean"),
    median_loan_amount=("loan_amount","median"),
    median_interest_rate=("interest_rate","median"),
).reset_index()

summary_gender
```

```python
summary_gender.to_csv("../outputs/summary_tables/summary_gender_2023_sample.csv", index=False)
```

```python
summary_state_gender = df.groupby(["state_code","applicant_gender"]).agg(
    applications=("approved","size"),
    approval_rate=("approved","mean")
).reset_index()

summary_state_gender.to_csv("../outputs/summary_tables/summary_state_gender_2023_sample.csv", index=False)
```
